```python
import requests

# Tu token de acceso
access_token = 'EAAW2PQ0k6ZC8BO7jQoYXzlzo3tu1hm9g1QnQYVAzCIMD9vek63EpMl0VeRYVxUweMqP1Vo4GrsdLViOEfXDKHhHsUQrYwUZCHDRTIaLij0nrYpfqZBwF7zmsGnSZAbUBAkMV3gOxmfaZBut5u3FRAm3XOyl2tk9KKgtsojUYUZA88P6HJcOgNyPwb7Hz78RUCZCm0GwEkRAJxZBSgTUZD'

# Endpoint de la API
url = 'https://graph.facebook.com/v20.0/120208378118290095/leads'

# Parámetros de la consulta
params = {
    'limit': 100,
    'access_token': access_token
}

# Realizar la solicitud GET
response = requests.get(url, params=params)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    data = response.json()
    print('Información de los Leads:')
    for lead in data['data']:
        dni = None
        correo = None
        contador=0
        for field in lead['field_data']:
            if contador==2:
                break
            if field['name'] == 'dni':
                dni = field['values'][0]
                contador+=1
            elif field['name'] == 'email':
                correo = field['values'][0]
                contador+=1
        print(f"DNI: {dni}, Correo: {correo}")
else:
    print(f"Error: {response.status_code}")
    print(response.json())
```
